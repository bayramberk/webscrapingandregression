```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import numpy as np
from datetime import datetime
import time
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    soup = bts(result.text, "html.parser")
    return soup
```

```python
## https://www.kitapyurdu.com/index.php?route=product/best_sellers&page=24&list_id=6&filter_in_stock=1&filter_in_stock=1&limit=100


pages = ["https://www.kitapyurdu.com/index.php?route=product/best_sellers&page=1&list_id=6&filter_in_stock=1&filter_in_stock=1&limit=100"] 
for page in range(2,6): 
    pages.append("https://www.kitapyurdu.com/index.php?route=product/best_sellers&page="+str(page)+"&list_id=6&filter_in_stock=1&filter_in_stock=1&limit=100")
pages
```

```python
product_links = []
for page in pages:
    html_pages = getAndParseURL(page)
    for product in html_pages.findAll("div",{"class":"name ellipsis"}):
        product_links.append(product.a.get("href"))
    time.sleep(5)    
```
